```python
tips=sb.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
# distplot
```

```python
sb.distplot(tips['total_bill'])
plt.show()
```

```python
# seaborn histogram
```

```python
sb.distplot(tips['total_bill'],kde=False)
plt.show()
```

```python
sb.distplot(tips['total_bill'],hist=False)       # kernsal plot
plt.show()
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips)          #jointplot 
plt.show()
             
```

```python
sb.scatterplot(x='total_bill',y='tip',data=tips)          #scatterplot 
plt.show()
```

```python
sb.set_style('ticks')         #set_style()                     
sb.pairplot(tips)             #pairplot()
plt.show()
```

```python
sb.stripplot(x='day',y='total_bill',data=tips)               #stripplot
plt.show()
```

```python
sb.stripplot(x='total_bill',y='tip',data=tips,jitter=True)     #stripplot,jitter
plt.show()
```

```python
sb.swarmplot(x='day',y='total_bill',data=tips)          #swarmplot 
plt.show()
```

```python
sb.boxplot(x=tips['total_bill'])                   # boxplot
plt.show()
```

```python
sb.boxplot(x='day',y='total_bill',data=tips)       # b0xplot
plt.show()
```

```python
sb.violinplot(x=tips['total_bill'])          #violin plot 
plt.show()
```

```python
sb.violinplot(x='day',y='total_bill',data=tips)          #jointplot 
plt.show()
```

```python
sb.pointplot(x='time',y='total_bill',data=tips)          #point plot 
plt.show()
```

```python
sb.regplot(x='total_bill',y='tip',data=tips)          #reg plot 
plt.show()
```
