```python
dat1={
    'name':['moon','sun','mars'],
    'distance':['8lak','120lak','65lak']
}
dat2={
    'name':['saturn','sun','jupiter'],
    'distance':['440lak','220lak','320lak']
}
DF1=pd.DataFrame(dat1)
DF2=pd.DataFrame(dat2)

newdf=DF1.merge(DF2,how='right')
newdf
```

```python
DF3=pd.merge(DF1,DF2,on='name',how='outer')
DF3
```

```python
DF3=pd.merge(DF1,DF2,on='name',how='right')
DF3
```

```python
DF3=pd.merge(DF1,DF2,on='name',how='left')
DF3
```
