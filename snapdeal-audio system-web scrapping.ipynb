```python
response=requests.get('https://www.snapdeal.com/products/electronics-home-audio-systems?sort=plrty')
bs=BeautifulSoup(response.content,'html.parser')
snapdeal_audio=bs.find_all('p', class_="product-title")

audio=[]
for i in range(0,len(snapdeal_audio)):
    audio.append(snapdeal_audio[i].get_text().strip())
audio
```

```python
response=requests.get('https://www.snapdeal.com/products/electronics-home-audio-systems?sort=plrty')
bs=BeautifulSoup(response.content,'html.parser')
snapdeal_offer_price=bs.find_all('span', class_='lfloat product-price')

offer_price=[]
for i in range(0,len(snapdeal_offer_price)):
    offer_price.append(snapdeal_offer_price[i].get_text().strip())
offer_price
```

```python
response=requests.get('https://www.snapdeal.com/products/electronics-home-audio-systems?sort=plrty')
bs=BeautifulSoup(response.content,'html.parser')
snapdeal_original_price=bs.find_all('span', class_='lfloat product-desc-price strike')

original_price=[]
for i in range(0,len(snapdeal_original_price)):
    original_price.append(snapdeal_original_price[i].get_text().strip())
original_price
```

```python
response=requests.get('https://www.snapdeal.com/products/electronics-home-audio-systems?sort=plrty')
bs=BeautifulSoup(response.content,'html.parser')
snapdeal_discount=bs.find_all('div', class_='product-discount')

discount=[]
for i in range(0,len(snapdeal_discount)):
    discount.append(snapdeal_discount[i].get_text().strip())
discount
```
